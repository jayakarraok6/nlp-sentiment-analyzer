```python
import textblob
from textblob import TextBlob

def analyze_sentiment(text):
    polarity = TextBlob(text).sentiment.polarity

    if polarity > 0:
        return "Positive"
    elif polarity < 0:
        return "Negative"
    else:
        return "Neutral"

review = input("Enter a review: ")

result = analyze_sentiment(review)

print("Sentiment:", result)
```
